## Algorithmic Flow 
At the heart of what we've been doing all semester is classifying a collection of observations into *groups* based on *features*. Each feature $F_i$ (including the bias term $F_0$) has an associated weight $w_i$ that classifies each observation into a group, using an algorithm given below. 


0. Assemble training data
    + Data can have any number of *features*, which we'll index by i.
    + Data can have any number of observations, or points, which we'll index by j.
    + Data must belong to one of many *classes,* which we'll index by k.
    + Randomize the order of the observation<p>
    <p>

1. Create a *Feature 0* for each observation
    + This is a *bias* term, esentially a constant
    + $F_{0,j}=1$
    + Or, in English, set Feature 0 equal to 1 for all observations $j=1, 2, 3...j_{max}$.<p>

2. Assign initial *weights,* one for each feature
    + Typically $w_i=U(-\sqrt{2/n}, +\sqrt{2/n})$, where n is the number of neurons in the layer.
    + Or, in English, set the weight for each feature $i=0, 1, 2...i_{max}$ to a random number drawn from a uniform distribution with mean 0 and a range of $\pm \sqrt{2/n}$. So, for $n=10$, $U(\approx \pm 0.45)$ <p>
    
3. Repeat the following loop until some "end" condition is reached, running through each data point in turn
    1. Locate the data point
        + find 
$$
\begin{align}
S_j=\sum w_i \cdot F_{i,j}
\end{align}
$$
        + Or, in English, find for each observation j: $S_j=w_0 \cdot F_{0,j}+w_1 \cdot F_{1,j}+w_2 \cdot F_{2,j}+... w_{i_{max}} \cdot F_{{i_{max}},j}$
        + This step converts the coordnates of a point to a "lever arm," which will help tilt the next generation of weights in a way that will tume them to seperate the groups.
    2. Predict which group the observation belongs to, by using a **activation function** based on the summed values of all the inputed S values. The activation function is simply a switch: if the sum of the inputs is that much, output this. 
    3. Change the weights, based on whether the observation was correctly classified.
        + If the *predicted* class is correct, weights are good, do nothing
        + But is the *predicted* class is wrong, change the weights in the direction that will improve next prediction
        + This is the AI, the learning step. 
        + $ \Delta w_i=r \cdot (C_j-P_j) \cdot F_{i,j}$
        + $r$ is the *learning rate*, where $0<r<1$. The rate has to be large enough to allow the loop to end, but small enough to prevent wild oscillations. Start with $r=0.1$
        
        + If the weights give a correct prediction, no change. But if they don't give the correct prediction, the weights are changed, and in a direction likely to lead to better prediction, by a proportion of $2r$ of the Feature value.

The activation function converts the sum of all the input neurons' signals into an output.
$$
\sum S_i \rightarrow \text{activation function} \rightarrow \text{Output}
$$

But how that conversion hapens matters. Recall that the whole point of AI is to label an unknown individual observation with a label using the features $F_i$. Multilayer systems work from Features, to a hidden layer, to classification, and then "back propagate" the results to adjust the weights. The activation function determines how signals are converted, ultimately, to labels.

The acivation function is conveniently shown by a function plot, with the sum of the inputs on the x axis, and the output on the y axis.


## Activation Function: A Natural History
### Duodecimal functions

The simplest activation function was used in the original perceptron:
 $$
\begin{align}
 O=\begin{cases}
    1, & \text{if $\sum S_i>0$}.\\
    -1, & \text{if $\sum S_i \leq 0$}.
  \end{cases}
\end{align}
        $$

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
x = np.arange(-4,4.1, .1)
y=np.sign(x)# plot data
plt.plot(x,y)

# plot data
plt.plot(x,y)
plt.grid(which='major', axis='both')
plt.xlabel('Sum of inputs, $\sum S_i$')
plt.ylabel('Output, $O$')
plt.title('Figure 1: The acivation function $O=sign(S_{total})$')

# plt.savefig('images/02_06.png', dpi=300)
plt.show()

```

This AF has three problems: it is discontinuous, it only has two states, and it *saturates* to a fixed output value ($\pm 1$). Imagine that you have four inputs, all of which are either 0 or 1. The step function will easily get pushed to  saturation. As Sharma notes

>Suppose you are creating a binary classifier. Something which should say a “yes” or “no” ( activate or not activate ). A Step function could do that for you! That’s exactly what it does, say a 1 or 0. Now, think about the use case where you would want multiple such neurons to be connected to bring in more classes. Class1, class2, class3 etc. What will happen if more than 1 neuron is “activated”. All neurons will output a 1 ( from step function). Now what would you decide? Which class is it? As noted by Brownlee,
>The simplest activation function is referred to as the linear activation, where no transform is applied at all. A network comprised of only linear activation functions is very easy to train, but cannot learn complex mapping functions. Linear activation functions are still used in the output layer for networks that predict a quantity (e.g. regression problems). 
Nonlinear activation functions are preferred as they allow the nodes to learn more complex structures in the data. Traditionally, two widely used nonlinear activation functions are the sigmoid and hyperbolic tangent activation functions.

## The Octavo functions
Non-linear activation functions allow more powerful, accurate and faster learning. In the 1990s and 2000s the sigmoid (technically called the error function, "erf") and hyperbolic tangent ("tanh") activation functions carried the load. The functions are quite similar in overall looks, but apparently the tanh is actually more efficient.

$$
\begin{align}
\text{Sigmoid: }& O=\frac{1}{1+e^{-S_{total}}}\\
\\
\text{tanh: }& O=\frac{2}{1+e^{-2S_{total}}}-1\\
\\
\text{tanh: }&  =2 \cdot \text{erf}(2S)-1
\end{align}
$$


```python
# plot data
y=1/(1+np.exp(-x))
yy=2/(1+np.exp(-2*x))-1
yyy=2/(1+np.exp(-x))-1
plt.plot(x,y,label='sigmoid')
plt.plot(x,yy, label='tanh')
plt.plot(x,yyy, label='scaled sig')
plt.grid(which='major', axis='both')
plt.xlabel('Sum of inputs, $\sum S_i$')
plt.ylabel('Output, $O$')
plt.title('Figure 5: The sigmoidal and tanh activation functions')
plt.legend(loc='upper left')
# 
plt.show()
```

Brownless again
> A general problem with both the sigmoid and tanh functions is that they saturate. This means that large [positive] values snap to 1.0 and [large negative] values snap to -1 or 0 for tanh and sigmoid respectively. Further, the functions are only really sensitive to changes around their mid-point of their input, such as 0.5 for sigmoid and 0.0 for tanh.

>The limited sensitivity and saturation of the function happen regardless of whether the summed activation from the node provided as input contains useful information or not. *Once saturated, it becomes challenging for the learning algorithm to continue to adapt the weights to improve the performance of the model.*

Note the inherent trade-offs in the AF: it can't saturate, it has to be steep, and it has to be continuous. No single function can optimally accomplish all of these simulataneously. So, choosing an AF contributes to the efficacy of the whole process.

### Rectified Linear Activation Function

Used primarily in hidden layers, the ReLU AF appears to be, well, the worst of all possible functions. 

```python
# plot data
y=np.maximum(0,x)

plt.plot(x,y,label='ReLu')

plt.grid(which='major', axis='both')
plt.xlabel('Sum of inputs, $\sum S_i$')
plt.ylabel('Output, $O$')
plt.title('Figure 6: The Rectified Linear AF')

# 
plt.show()
```

But empirically it appears to work effectively. The ReLu is efficient and well tested. Use it on hidden layers in most situations. Brownlee notes...
>The rectified linear activation function has rapidly become the default activation function when developing most types of neural networks.

Three aspects of the ReLu are advantageous, *via* Brownlee:
1. Computational Simplicity. The rectifier function is trivial to implement, requiring a max() function. This is unlike the tanh and sigmoid activation function that require the use of an exponential calculation. (Solution time reduced.)
2. Representational Sparsity. An important benefit of the rectifier function is that it is capable of outputting a true zero value. Unlike the tanh and sigmoid AF, ReLu gives a 0, not 1.234e-7. This means that negative inputs can output true zero values allowing the activation of hidden layers in neural networks to contain one or more true zero values. This is called a sparse representation and is a desirable property in representational learning as it can accelerate learning and simplify the model.
3. Linear Behavior. The rectifier function mostly looks and acts like a linear activation function. In general, a neural network is easier to optimize when its behavior is linear or close to linear.

I *think* that the linearity here is actually about the lack of saturation...there is a gradient regardless of $S_{total}$, as long as $S_{total}>0$, which allows updating of weights. The complete lack of gradient at $S_{total}<0$ is, I think, a way of "killing" neurons that are negative, which reduces their influence.

### SoftMax
Another branded function. In essence, SoftMax is another linear/sigmoid. Yeah, I know, we just learned that a linear function is best. And it requires not one but two exponentiations! Recall what Dr. Merwin said last week: little progress in 20 years. The sort of self-contradictory behavior is the hall mark of a field in which empirical results drive learning. That's not a bad thing, just an indication that the field still lacks a coherent, theory-based foundation. 

```python
def softmax(z):
    """Compute softmax values for each sets of scores in x."""
    return np.exp(z) / np.sum(np.exp(z), axis=0) 
```

```python

# plot data

y=np.random.normal(size=np.size(x, axis=0))
y=np.cumsum(softmax(y))

plt.plot(x,y,label='Softmax')

plt.grid(which='major', axis='both')
plt.xlabel('Sum of inputs, $\sum S_i$')
plt.ylabel('Output, $O$')
plt.title('Figure 7: The SoftMax AF')

# 
plt.show()
```
